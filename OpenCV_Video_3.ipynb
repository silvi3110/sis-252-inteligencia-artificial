```python
!pip install opencv-python opencv-python-headless

import cv2
from google.colab.patches import cv2_imshow

cap = cv2.VideoCapture('Gente caminando.mp4')
fgbg = cv2.createBackgroundSubtractorMOG2()
line_position = 400
conteo = 0

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    frame_resized = cv2.resize(frame, (640, 480))
    mask = fgbg.apply(frame_resized)
    _, mask = cv2.threshold(mask, 200, 255, cv2.THRESH_BINARY)
    contours, _ = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    cv2.line(frame_resized, (0, line_position), (640, line_position), (255, 0, 0), 2)

    for cnt in contours:
        area = cv2.contourArea(cnt)
        if area > 500:
            x, y, w, h = cv2.boundingRect(cnt)
            centro_y = y + h // 2
            if line_position - 2 < centro_y < line_position + 2:
                conteo += 1
            cv2.rectangle(frame_resized, (x, y), (x+w, y+h), (0, 255, 0), 2)

    cv2.putText(frame_resized, f"Conteo: {conteo}", (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2)
    cv2_imshow(frame_resized)

cap.release()
cv2.destroyAllWindows()
```
