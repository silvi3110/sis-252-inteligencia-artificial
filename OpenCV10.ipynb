```python
import cv2
import matplotlib.pyplot as plt
```

```python
# Cargar clasificador preentrenado de rostros
detector_rostros = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
```

```python
# Leer imagen
imagen = cv2.imread('personas.jpg')
gris = cv2.cvtColor(imagen, cv2.COLOR_BGR2GRAY)
```

```python
# Detectar rostros - scaleFactor controla escalado, minNeighbors precisión
rostros = detector_rostros.detectMultiScale(gris, scaleFactor=1.1, minNeighbors=5)
```

```python
# Dibujar rectángulos alrededor de rostros detectados
for (x, y, w, h) in rostros:
    cv2.rectangle(imagen, (x, y), (x+w, y+h), (0, 255, 0), 50)
```

```python
# Mostrar resultado
plt.imshow(cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB))
plt.title('Rostros Detectados')
plt.axis('off')
plt.show()
```
