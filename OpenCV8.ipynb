```python
import cv2
import matplotlib.pyplot as plt
```

```python
# Leer imagen en escala de grises
imagen = cv2.imread('imagenes.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
# Aplicar suavizado para reducir ruido antes de detectar bordes
imagen_suavizada = cv2.GaussianBlur(imagen, (5, 5), 1.4)
```

```python
# Detección de bordes usando el algoritmo de Canny
# Parámetros: umbral inferior, umbral superior
bordes = cv2.Canny(imagen_suavizada, 50, 150)
```

```python
# Mostrar resultados
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title('Imagen Original')
plt.imshow(imagen, cmap='gray')

plt.subplot(1, 2, 2)
plt.title('Bordes Detectados')
plt.imshow(bordes, cmap='gray')
plt.show()
```
